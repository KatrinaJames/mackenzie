```python
import pandas as pd
import statsmodels.api as sm
import yfinance as yf
```

```python
etf_factors = ['MTUM', 'QUAL', 'VLUE', 'SIZE', 'USMV']
data = yf.download(etf_factors, start='2024-03-14', end='2025-03-15')
data = data.loc[:,('Close', slice(None))]
data.columns = etf_factors
data
```

```python
etf_returns = data.pct_change().dropna()
etf_returns.index = pd.to_datetime(pd.Series(etf_returns.index)).dt.date
etf_returns
```

```python
cig_returns = pd.read_csv('CIG4778-prices.csv')[['Date', 'NAVPS']]
cig_returns['Date'] = pd.to_datetime(cig_returns['Date']).dt.date
cig_returns = cig_returns.sort_values(by='Date', ascending=True).set_index('Date').pct_change().dropna().rename(columns={"NAVPS": "CIG4778"})
cig_returns
```

```python
dyn_returns = pd.read_csv('DYN426-prices.csv')[['Price', 'Closing Date']]
dyn_returns['Closing Date'] = pd.to_datetime(dyn_returns['Closing Date']).dt.date
dyn_returns = dyn_returns.sort_values(by='Closing Date', ascending=True).rename(columns={"Closing Date": "Date", "Price": "DYN426"}).dropna().set_index('Date').pct_change().dropna()
dyn_returns
```

```python
fid_returns = pd.read_csv('FID1640-prices.csv')[['NAV', 'Date']]
fid_returns['Date'] = pd.to_datetime(fid_returns['Date']).dt.date
fid_returns = fid_returns.sort_values(by='Date', ascending=True).set_index('Date').pct_change().dropna().rename(columns={"NAV": "FID1640"})
fid_returns
```

```python
cig_returns = pd.merge(cig_returns, etf_returns, on='Date', how='outer').dropna()
cig_returns
```

```python
fid_returns = pd.merge(fid_returns, etf_returns, on='Date', how='outer').dropna()
fid_returns
```

```python
dyn_returns = pd.merge(dyn_returns, etf_returns, on='Date', how='outer').dropna()
dyn_returns
```

```python
cig_regression = sm.OLS(cig_returns['CIG4778'], sm.add_constant(cig_returns[etf_factors])).fit()
cig_regression.params
```

```python
fid_regression = sm.OLS(fid_returns['FID1640'], sm.add_constant(fid_returns[etf_factors])).fit()
fid_regression.params
```

```python
dyn_regression = sm.OLS(dyn_returns['DYN426'], sm.add_constant(dyn_returns[etf_factors])).fit()
dyn_regression.params
```
